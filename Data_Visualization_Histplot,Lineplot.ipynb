```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
sns.histplot(data=df,x='age')
```

```python
sns.histplot(data=df,y='age')
```

```python
sns.barplot(data=df,x='age',bins=10)
```

```python
sns.barplot(data=df,x='age',kde=True)
plt.grid(True)
```
