```python
import numpy as np
import matplotlib.pyplot as plt

n=10
p=0.5
size=1000
data=np.random.binomial(n,p,size)
plt.hist(data,bins=range(n+2),density=True,alpha=0.5,color="g",edgecolor="black")
plt.title('binomial distribution(n=10,p=0.5)')
plt.xlabel('number of successes')
plt.ylabel('probability')
plt.show()
```

```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt

lambda_val = 5
size=1000
data = np.random.poisson(lambda_val, size)
n = max(data)
plt.hist(data,bins=range(min(data),max(data)+1), density=True, alpha=0.5, color="g", edgecolor="black")
plt.title('poisson distribution(sigma=2)')
plt.xlabel('number of occurrences')
plt.ylabel('probability')
plt.show()
```

```python
lambda_val= 2
size = 1000
data = np.random.exponential(1/lambda_val, size)
plt.hist(data, bins=50, density=True, alpha=0.6, color='b', edgecolor='black')
xmin, xmax = plt.xlim()
from scipy.stats import expon
x = np.linspace(xmin, xmax, 100)
y = expon.pdf(x, scale=1/lambda_val)
plt.plot(x, y, 'r', lw=2)
plt.title('Exponential Distribution')
plt.xlabel('Random Variable (x)')
plt.ylabel('Probability Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
mu = 0
sigma = 1
size = 1000
data = np.random.normal(mu, sigma, size)
plt.hist(data, bins=50, density=True, alpha=0.6, color='b', edgecolor='black')
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = stats.norm.pdf(x, mu, sigma)
plt.plot(x, y)
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.show()
```
