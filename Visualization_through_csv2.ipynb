```python
! pip install pandas
import pandas as pd
data=pd.read_csv("Dream.csv")
```

```python
data.head(10)
```

```python
sum_data=data["Loan Amount"].sum()
print(sum_data)

mean_data=data['Loan Amount'].mean()
print(mean_data)

median_data=data['Loan Amount'].median()
print(median_data)

mode_data=data['Loan Amount'].mode()
print(mode_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.barplot(x="Loan Eligibility",y="Loan Amount",data=data)
plt.show()
```

```python
sns.barplot(x="Loan Amount",y="Loan Eligibility",data=data)
plt.show()
```

```python
sns.barplot(x="Loan Eligibility",y="Loan Amount",hue="Gender",data=data)
plt.show()
```

```python
sns.barplot(x="Loan Amount",y="Loan Eligibility",hue="Gender",data=data)
plt.show()
```

```python
sns.barplot(x="Loan Eligibility",y="Loan Amount",hue="Gender",palette='pastel',data=data)
plt.show()
```
