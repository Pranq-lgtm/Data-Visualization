```python
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
```

```python
iris.head(5)
```

```python
mean=iris['sepal_length'].mean()
mode=iris['sepal_length'].mode()
```

```python
print("mean:",mean)
print("mode:",mode)
```
